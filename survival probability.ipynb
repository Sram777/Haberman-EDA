```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\saira\Downloads\haberman.csv")
df.head()
```

```python
print(df.status.unique())
```

```python
#Univariate Analysis

sns.FacetGrid(df,hue='status',height=5).map(sns.distplot,'age').add_legend()
sns.FacetGrid(df,hue='status',height=5).map(sns.distplot,'year').add_legend()
sns.FacetGrid(df,hue='status',height=5).map(sns.distplot,'nodes').add_legend()
```

```python
status_1=df.loc[df['status']==1]
```

```python
plt.figure(figsize=(20,6))
plt.subplot(131)
counts,bin_edges=np.histogram(status_1["age"],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,linewidth=3.0)
plt.plot(bin_edges[1:],cdf,linewidth=3.0)

plt.subplot(132)
counts,bin_edges=np.histogram(status_1["nodes"],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,linewidth=3.0)
plt.plot(bin_edges[1:],cdf,linewidth=3.0)

plt.subplot(133)
counts,bin_edges=np.histogram(status_1["year"],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,linewidth=3.0)
plt.plot(bin_edges[1:],cdf,linewidth=3.0)

```
